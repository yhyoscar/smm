```python
# statistics of log-return
quote = getonequote('GOOG')
quote = add_logreturn(quote, range(6))
for k in range(6):
    quote['logreturn_'+format(k)] *= 100
quote.head(5)
```

```python
from scipy.stats import norm

# histogram of log-return
plt.figure(1, figsize=(15,6))
for k in range(6):
    plt.subplot(2,3,k+1)
    x = quote['logreturn_'+format(k)].dropna()
    mu, std = norm.fit(x)
    xnew = np.linspace(min(x), max(x), 1000)
    fitpdf = norm.pdf(xnew, mu, std)
    
    plt.hist(x, bins=200, normed=True)
    plt.plot(xnew, fitpdf, color='r', lw=2)
    plt.title('GOOG: logreturn_'+format(k)+ \
              ' (mu='+format(mu,'.2f')+'%, std='+format(std,'.2f')+'%)', loc='left', fontsize=12)
#log(x(t)) - log(x(t-1))
#= log (1 + r) ~ r+r^2+r^3+...
plt.tight_layout()
```

```python
ss = pd.read_csv(fslist)
capnum = [capstr2num(x) for x in ss['MarketCap'].values ]
ss['MarketCap'] = capnum
stop = ss.sort_values('MarketCap', ascending=False).head(10)
stop
```

```python
quotes = getmulquote(stop['Symbol'].values)
for s in quotes.keys():
    print('-------- ', s, ' --------')
    quotes[s] = add_logreturn(quotes[s], [0, 1, 2])
```

```python
comps = ['AAPL', 'GOOG']
sall = pd.merge(quotes[comps[0]], quotes[comps[1]], on=['Date'], 
                suffixes=['_'+comps[0], '_'+comps[1]])
sall.head(5)
```

```python
plt.figure(1, figsize=(15,5))
for i in range(3):
    plt.subplot(1,3,i+1)
    x = sall[['logreturn_'+format(i)+'_'+comps[0], 'logreturn_'+format(i)+'_'+comps[1]] ].dropna()
    r, p = pearsonr(x.iloc[:,0], x.iloc[:,1])
    plt.plot(x.iloc[:,0], x.iloc[:,1], '.')
    plt.title('logreturn_'+format(i)+': r='+format(r,'.2f')+', p='+format(p,'.2f'), loc='left')
    plt.xlabel(comps[0])
    plt.ylabel(comps[1])
plt.tight_layout()    
```
